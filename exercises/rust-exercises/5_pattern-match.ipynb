# Match

1.

```rust
// Fill the blanks
enum Direction {
    East,
    West,
    North,
    South,
}
 
fn main() {
    let dire = Direction::South;
    match dire {
        Direction::East => println!("East"),
        Direction::South | Direction::North  => { // Matching South or North here
            println!("South or North");
        },
        _ => println!("West"),
    };
}
 
main();
```

2. Match is an expression, so we can use it in assignments

```rust
fn main() {
    let boolean = true;
 
    // Fill the blank with a match expression:
    //
    // boolean = true => binary = 1
    // boolean = false =>  binary = 0
   
    // let binary = boolean as i32;
    let binary = match boolean {
        false => 0,
        true => 1,
    };
 
    assert_eq!(binary, 1);
 
    println!("Success!");
}
 
 main();
```

3. Using match to get the data an enum variant holds.

```rust
// Fill in the blanks
#[derive(Debug)]
enum Message {
    Quit,
    Move { x: i32, y: i32 },
    Write(String),
    ChangeColor(i32, i32, i32),
}

fn main() {
    let msgs = [
        Message::Quit,
        Message::Move{x:1, y:3},
        Message::ChangeColor(255,255,0)
    ];

    for msg in msgs {
        show_message(msg)
    }

    println!("Success!");
} 

fn show_message(msg: Message) {
    match msg {
        Message::Move{x, y} => { // match  Message::Move
            assert_eq!(x, 1);
            assert_eq!(y, 3);
        },
        Message::ChangeColor(r, g, b) => {
            assert_eq!(g, 255);
            assert_eq!(b, 0);
        },
        Message::Write(text) => {
            println!("Text message: {}", text);
        },
        Message::Quit => println!("no data in these variants")
    }
}

main();
```
